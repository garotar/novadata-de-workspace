```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
np.random.seed(42)

response_times = np.random.lognormal(mean=np.log(100), sigma=0.4, size=10000)

plt.figure(figsize=(10,4))
sns.histplot(response_times, bins=100, kde=True)
plt.xlabel('Время отклика')
plt.ylabel('Количество')
plt.show()
```

```python
shapiro_stats, shapiro_pvalue = stats.shapiro(response_times)
print(f'p_value = {shapiro_pvalue:.5f} меньше уровня значимости 0.05 (5%), следовательно, отвергаем нулевую гипотезу о том, что данные пришли из нормального распределения')
```

```python
res = stats.anderson(response_times, dist='norm')
```

```python
log_response_times = np.log(response_times)

plt.figure(figsize=(10,4))
sns.histplot(log_response_times, bins=100, kde=True)
plt.xlabel('Время отклика')
plt.ylabel('Количество')
plt.show()
```

```python
shapiro_stats2, shapiro_pvalue2 = stats.shapiro(log_response_times)
print(f'p_value = {shapiro_pvalue2:.5f} сильно больше уровня значимости 0.05 (5%), следовательно, не можем отвргать гипотезу о том, что распределение - нормальное')
```
